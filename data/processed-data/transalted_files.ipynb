```python
import pandas as pd
import os

# ==============================================================================
#1. CLIENTS
# ==============================================================================

#Translation of Clients file
clients = "Clients.xlsx"

if os.path.exists(clients):
    df_clients = pd.read_excel(clients, header=0)
    print(f"The file {clients} is uploaded successfully!")
    display(df_clients.head(10))
else:
    print(f"ERROR: The file is not found {clients}. Make sure to upload the file to google colab.")
    print("Available folders:", [f for f in os.listdir('.') if os.path.isfile(f)])
```

```python
# Clients Dictionary: Polish column name → English column name

df_clients = df_clients.rename(columns ={
    "ID klienta": "client_id",
    "Nazwa": "name",
    "Miasto": "city",
    "Branża": "industry"
})


# Translate text values in the "industry" column
# Map: Polish value → English value
industry_map = {
    "IT":       "IT",        # unchanged
    "Retail":   "Retail",    # unchanged
    "Finanse":  "Finance",
}
df_clients["industry"] = df_clients["industry"].map(industry_map)
df_clients.head()
```

```python
# Save the translated file
df_clients.to_csv("clients_translated.csv", index=False, encoding="utf-8-sig")

# Download it to your computer
from google.colab import files
files.download("clients_translated.csv")
```

```python
# ==============================================================================
#2.COSTS
# ==============================================================================

#Upload of Costs file
costs = "Costs.csv"

if os.path.exists(costs):
    df_costs = pd.read_csv(costs, header = 1)
    print(f"The file {costs} is uploaded successfully!")
    display(df_costs.head(10))
else:
    print(f"ERROR: The file is not found {costs}. Make sure to upload the file to google colab.")
    print("Available folders:", [f for f in os.listdir('.') if os.path.isfile(f)])
```

```python
# Costs Dictionary: Polish column name → English column name

#removing space in the headers
df_costs.columns = df_costs.columns.str.strip()

#translation
df_costs = df_costs.rename(columns ={
    "Data": "date",
    "Projekt": "project",
    "Koszt": "cost",
})

#PLN removal from costs values
df_costs['cost'] = df_costs['cost'].str.replace('PLN ', '', regex = False)
df_costs.head()
```

```python
# Save the translated file
df_costs.to_csv("costs_translated.csv", index=False, encoding="utf-8-sig")

# Download it to laptop
from google.colab import files
files.download("costs_translated.csv")
```

```python
# ==============================================================================
#3.INVOICES
# ========================================================================

#Upload of Invoices file
invoices = "Invoices.csv"

if os.path.exists(invoices):
    df_invoices = pd.read_csv(invoices, sep=";", header = 0, encoding="windows-1250")
    print(f"The file {invoices} is uploaded successfully!")
    display(df_invoices.head(10))
else:
    print(f"ERROR: The file is not found {invoices}. Make sure to upload the file to google colab.")
    print("Available folders:", [f for f in os.listdir('.') if os.path.isfile(f)])
```

```python
# Invoices Dictionary: Polish column name → English column name

#removing space in the headers
df_invoices.columns = df_invoices.columns.str.strip()

#translation
df_invoices = df_invoices.rename(columns ={
    "Nr faktury": "invoice_id",
    "Projekt": "project",
    "Wartość netto": "net_value",
    "Data wystawienia": "invoice_date"
})

#zł removal from costs values
df_invoices['net_value'] = df_invoices['net_value'].str.replace(' zł', '', regex = False)

df_invoices.head()
```

```python
# Save the translated file
df_invoices.to_csv("invoices_translated.csv", index=False)

# Download it to laptop
from google.colab import files
files.download("invoices_translated.csv")
```

```python
# ==============================================================================
#4.PROJECTS
# ========================================================================
#Upload of Projects file
projects = "Projects.csv"

if os.path.exists(projects):
    df_projects = pd.read_csv(projects, sep=";", header = 0, encoding="windows-1250")
    print(f"The file {projects} is uploaded successfully!")
    display(df_projects.head())
else:
    print(f"ERROR: The file is not found {projects}. Make sure to upload the file to google colab.")
    print("Available folders:", [f for f in os.listdir('.') if os.path.isfile(f)])
```

```python
# Projects Dictionary: Polish column name → English column name

#removing space in the headers
df_projects.columns = df_projects.columns.str.strip()

#translation
df_projects.columns = ["project_id", "client_id", "project_type"]
df_projects.head()
```

```python
print(df_projects["project_type"].unique())

df_projects["project_type"] = df_projects["project_type"].str.strip()

project_type_map = {
    "Wdrożenie ERP":        "erp_implementation",
    "BI":                    "BI",    # unchanged
    "Usługi szkoleniowe":  "training_services",
}

df_projects["project_type"] = df_projects["project_type"].map(project_type_map)
df_projects.head()
```

```python
# Save the translated file
df_projects.to_csv("projects_translated.csv", index=False)

# Download it to laptop
from google.colab import files
files.download("projects_translated.csv")
```
